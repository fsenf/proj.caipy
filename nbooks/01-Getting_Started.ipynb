# Segmentation Example

Below, I provide a very simple segmentation example. I use the `segmentation.py` functionality that is in my `analysis_tools` toolbox of `tropy`. The `PYTHONPATH` is by default directly poiting inside tropy, hence only relative module structure has to be imported. 

````python
%matplotlib inline
import analysis_tools.segmentation as seg
import numpy as np
import pylab as pl

pl.rcParams['figure.figsize'] = (10.0, 8.0)
pl.rcParams['font.size'] = 16.0
````

I use the `matplotlib` library `pylab` to get the stuff visualized. In the below example, 
* normal-distributed random number are drawn, 
* a threshold of 0.1 is applied
* a segmentation technique is used to generate cells.
* background (= 0) is masked for nicer plot

````python
f = np.random.randn(100,100)
c = seg.clustering(f, 0.1)
cm = np.ma.masked_equal(c, 0)

pl.imshow(cm , interpolation = 'nearest')
pl.title('A simple Segmentation Example')
````

One step, for more realistic data fields.

I use Gaussian smoothing from the `scipy.ndimage` library.


````python
import scipy.ndimage

f_smooth = scipy.ndimage.gaussian_filter(f, 2)

c = seg.clustering(f_smooth, 0.1)
cm = np.ma.masked_equal(c, 0)

pl.imshow(cm , interpolation = 'nearest')
pl.title('A smoother Segmentation Example')
````
